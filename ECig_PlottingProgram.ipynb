```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Unflavored_Plot(mean_cigarette_uf,mean_savory_uf,mean_sweet_uf):
    ##This function takes the inputs of the mean cigarette, savory, and sweet cravings for an unflavored E-cig user
    ##It then plots these craving scores and assigns labels to the different parts of the graph.
    mean_cigarette_uf = [1,2,3,4,5,6,7]
    mean_savory_uf = [7,6,5,4,3,2,1]
    mean_sweet_uf = [1,3,2,4,5,7,6]
    duration=np.arange(7)
    bounds = [0,len(duration)+1,0,7]
    width=0.25
    bar1=np.arange(len(duration))
    bar2=[i+width for i in bar1]
    bar3=[i+width for i in bar2]
    plt.bar(bar1, mean_cigarette_uf, width, align="edge", label="Cigarettes")
    plt.bar(bar2, mean_savory_uf, width, align="edge", label="Savory")
    plt.bar(bar3, mean_sweet_uf, width, align="edge", label="Sweet")
    plt.axis(bounds)
    #The logistics of the function, which displays a legend and assigns axis and chart labels
    plt.title("Unflavored E-Cig Results")
    plt.ylabel("Craving Level")
    plt.xlabel("Days Since Last Use")
    plt.legend(ncol=1, loc="best")
    plt.show()
```

```python
def Flavored_Plot(mean_cigarette_f,mean_savory_f,mean_sweet_f):
    ##This function takes the inputs of the mean cigarette, savory, and sweet cravings for a flavored E-cig user
    ##It then plots these craving scores and assigns labels to the different parts of the graph.
    duration=np.arange(7)
    bounds = [0,len(duration)+1,0,7]
    width=0.25
    mean_cigarette_f = [1,2,3,4,5,6,7]
    mean_savory_f = [7,6,5,4,3,2,1]
    mean_sweet_f = [1,3,2,4,5,7,6]
    bar1=np.arange(len(duration))
    bar2=[i+width for i in bar1]
    bar3=[i+width for i in bar2]
    plt.bar(bar1, mean_cigarette_f, width, align="edge", label="Cigarettes")
    plt.bar(bar2, mean_savory_f, width, align="edge",label="Savory")
    plt.bar(bar3, mean_sweet_f, width,  align="edge", label="Sweet")
    plt.axis(bounds)
    #The logistics of the function, which displays a legend and assigns axis and chart labels
    plt.title("Flavored E-Cig Results")
    plt.ylabel("Craving Level")
    plt.xlabel("Days Since Last Use")
    plt.legend(ncol=1, loc="best")
    plt.show()
```

```python
mean_cigarette_uf = [1,2,3,4,5,6,7]
mean_savory_uf = [7,6,5,4,3,2,1]
mean_sweet_uf = [1,3,2,4,5,7,6]
```

```python
Unflavored_Plot(np.arange(7),np.arange(7),np.arange(7))
```

```python
Flavored_Plot(np.arange(7),np.arange(7),np.arange(7))
```
